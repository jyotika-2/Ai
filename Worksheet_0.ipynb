# **Problem - 4: Matrix Operations:**
**• For the following arrays:**

**A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);**

**Prove following with Numpy:**

**1. Prove A.A−1 = I.**

```python
import numpy as np

# Given matrix A
A = np.array([[3, 4], [7, 8]])

# Calculate inverse A⁻¹
A_inv = np.linalg.inv(A)

# Compute A · A⁻¹
result = np.dot(A, A_inv)

# Identity matrix I (2x2)
I = np.eye(2)

print("A:\n", A)
print("A⁻¹:\n", A_inv)
print("A · A⁻¹:\n", result)
print("Is equal to I?", np.allclose(result, I))
```

**2. Prove AB ̸= BA.**

```python
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Compute AB and BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("A:\n", A)
print("B:\n", B)
print("AB:\n", AB)
print("BA:\n", BA)
print("AB == BA?", np.array_equal(AB, BA))
```

**3. Prove (AB) T = BTAT.**



```python
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Compute both sides
AB = np.dot(A, B)
AB_transpose = np.transpose(AB)

BT_AT = np.dot(np.transpose(B), np.transpose(A))
# or: np.transpose(A) @ np.transpose(B)

print("AB:\n", AB)
print("(AB)ᵀ:\n", AB_transpose)
print("BᵀAᵀ:\n", BT_AT)
print("Equal?", np.array_equal(AB_transpose, BT_AT))
```

**• Solve the following system of Linear equation using Inverse Methods.**

**2x − 3y + z = −1**

**x − y + 2z = −3**

**3x + y − z = 9**

**{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}**


```python
import numpy as np

# Coefficient matrix A and constant vector B
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

# Method 1: X = A⁻¹B (using inverse)
A_inv = np.linalg.inv(A)
X_inverse = np.dot(A_inv, B)

print("Solution using A⁻¹B:")
print("x =", X_inverse[0])
print("y =", X_inverse[1])
print("z =", X_inverse[2])
print("X =", X_inverse)
```

**• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function of Numpy}**


```python
import numpy as np

# Matrix form: AX = B
A = np.array([[2, -3,  1],
              [1, -1,  2],
              [3,  1, -1]])

B = np.array([-1, -3, 9])

# Step 1: Compute inverse using np.linalg.inv()
A_inv = np.linalg.inv(A)
print("A⁻¹ =\n", A_inv)

# Step 2: Solve X = A⁻¹B
X = np.dot(A_inv, B)
print("\nSolution X =", X)
print("x =", X[0], "y =", X[1], "z =", X[2])
```

# **10.2 Experiment: How Fast is Numpy?**

**In this exercise, you will compare the performance and implementation of operations using plain Python
lists (arrays) and NumPy arrays.**

**Follow the instructions:**

**1. Element-wise Addition:**

**• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.**




```python
import time

# Create two lists of 1,000,000 elements
size = 1000000
list1 = list(range(size))
list2 = list(range(size, size + size))

# Measure time for element-wise addition using list comprehension
start_time = time.time()

# Element-wise addition: [a + b for each pair]
result_list = [list1[i] + list2[i] for i in range(size)]

list_time = time.time() - start_time

print(f"Size: {size:,} elements")
print(f"Python Lists - Element-wise addition time: {list_time:.4f} seconds")
print(f"Result preview: {result_list[:5]}...")  # First 5 elements
```

**• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.**

```python
import numpy as np
import time

# Create two NumPy arrays of 1,000,000 elements
size = 1000000
array1 = np.arange(size)
array2 = np.arange(size, size + size)

# Measure time for element-wise addition
start_time = time.time()

# Single operation: vectorized element-wise addition
result_array = array1 + array2

numpy_time = time.time() - start_time

print(f"Size: {size:,} elements")
print(f"NumPy Arrays - Element-wise addition time: {numpy_time:.4f} seconds")
print(f"Result preview: {result_array[:5]}...")  # First 5 elements
```

**4. Matrix Multiplication**

**• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.**


```python
import time
import random

# Create two 1000x1000 matrices
n = 1000
matrix1 = [[random.randint(0, 10) for _ in range(n)] for _ in range(n)]
matrix2 = [[random.randint(0, 10) for _ in range(n)] for _ in range(n)]

# Measure time for matrix multiplication
start_time = time.time()

# Triple nested loop matrix multiplication: C[i][j] = Σ(A[i][k] * B[k][j])
result = [[0 for _ in range(n)] for _ in range(n)]
for i in range(n):
    for j in range(n):
        for k in range(n):
            result[i][j] += matrix1[i][k] * matrix2[k][j]

python_time = time.time() - start_time

print(f"Matrix size: {n}×{n} = {n*n:,} elements")
print(f"Python Lists - Matrix multiplication time: {python_time:.2f} seconds")
print(f"Result preview (top-left 3×3):\n{result[:3][:3]}")
```

**• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.**

```python
import numpy as np
import time

# Create two 1000x1000 random matrices
n = 1000
matrix1 = np.random.rand(n, n)
matrix2 = np.random.rand(n, n)

# Measure time for matrix multiplication
start_time = time.time()

# Single operation: optimized BLAS matrix multiplication
result = np.dot(matrix1, matrix2)
# Alternative: result = matrix1 @ matrix2

numpy_time = time.time() - start_time

print(f"Matrix size: {n}×{n} = {n*n:,} elements")
print(f"NumPy Arrays - Matrix multiplication time: {numpy_time:.2f} seconds")
print(f"Result shape: {result.shape}")
print(f"Result preview (top-left 3×3):\n{result[:3, :3]}")
```
